```python
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS

model = ChatterboxTTS.from_pretrained(device="cuda")

text = "Happy Birthday! Wishing you a wonderful day full of joy and laughter. May this year bring you success, good health, and happiness. Celebrate big — you deserve all the best life has to offer!"
# wav = model.generate(text)
# ta.save("test-1.wav", wav, model.sr)

# If you want to synthesize with a different voice, specify the audio prompt
AUDIO_PROMPT_PATH="/content/Chandan.wav"
wav = model.generate(text, audio_prompt_path=AUDIO_PROMPT_PATH)
ta.save("PayelSuthar-BirthdayWish.wav", wav, model.sr)
```
